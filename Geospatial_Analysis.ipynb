```
import geopandas as gpd
```

# Basic World Map

```
gpd.datasets.available
```

```
world_filepath = gpd.datasets.get_path('naturalearth_lowres')
world = gpd.read_file(world_filepath)
world.head()
```

```
city_filepath = gpd.datasets.get_path('naturalearth_cities')
city = gpd.read_file(city_filepath)
city.head()
```

```
ax = world.plot(figsize=(20,20), color='whitesmoke', edgecolor='black')
city.plot(ax=ax, markersize=30)
```
